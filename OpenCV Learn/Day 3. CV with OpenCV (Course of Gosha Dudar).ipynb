```python
# Get video from user's web camera

video = cv2.VideoCapture(0)
video.set(3, 500)  # width is '3'
video.set(4, 300)  # height is '4'

while 'OpenCV':
    success, img = video.read()
    cv2.imshow('Frame', img)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cv2.destroyAllWindows()  # To close window
```

```python
# To switch off user's camera
video.release()
```

```python
# My class 'Webcam'
import cv2


# Class for user's image by webcam
class Webcam:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.webcam_on = None
        
    def show_user(self):
        video = cv2.VideoCapture(0)
        video.set(3, self.width)
        video.set(4, self.height)
        
        self.webcam_on = True
        while 'OpenCV is cool':
            _success, _img = video.read()
            cv2.imshow('Frame', _img)
            
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
                
    def close_webcam(self):
        if self.webcam_on:
            video.release()
            cv2.destroyAllWindows() 
```

```python
# My object 'film'
film = Webcam(1920, 1080)
film.show_user()
film.close_webcam()
```

```python
# Uded ChatGPT
import cv2

class VideoCaptureManager:
    def __init__(self, width=500, height=300):
        self._video = cv2.VideoCapture(0)
        self._video.set(3, width)  # Set width
        self._video.set(4, height)  # Set height

    def run(self):
        while True:
            success, img = self._video.read()
            cv2.imshow('Frame', img)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

    def close(self):
        self._video.release()
        cv2.destroyAllWindows()

if __name__ == "__main__":
    manager = VideoCaptureManager()
    manager.run()
    manager.close()
```

```python
# Import libraries
import numpy as np
import cv2
```

```python
# Load an image
img = cv2.imread('images/woman.jpg')
```

```python
# Show the image for 10 seconds
cv2.imshow('Woman', img)
cv2.waitKey(10_000)
cv2.destroyAllWindows()
```

```python
# Shape of the image
img.shape
```

The terminal outputs a tuple of three elements (height, width, number of layers), 3 layers - RGB

```python
# Resize the image disproportionately

# img = cv2.resize(img, (300, 500))
```

```python
# Resize the image Proportionately
height = img.shape[0]
weight = img.shape[1]
img = cv2.resize(img, (height // 9, weight // 9))
```

```python
# Show the image forever

cv2.imshow('Woman', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import pandas as pd
```

```python
ign = pd.read_csv('ign.csv', index_col='Unnamed: 0')
ign.drop(columns='url', inplace=True)
ign.head()
```

```python
new_indexes = ign.index
new_indexes
```

```python
ign.sort_values(by=['score', 'title'], ascending=[False, True], inplace=True)
ign.head()
```

```python
ign.index
```

```python
result = ign[['title', 'score', 'score_phrase', 'platform', 'genre', 'release_year', 'release_month', 'release_day']]
result.index = new_indexes
result.head()
```

```python
result.to_csv(path_or_buf='IGN games from best to worst.csv', index=False, sep=',')
```

#  Основы компьютерного зрения: мир глазами машины

В синие нейроны - это входы, в них нет весов, а вот каждое ребро имеет вес. 4 * 4 = 16, 4 * 5 = 20, 5 * 6 = 30, 6 * 4 = 24, 4 * 3 = 12 и 3 * 3 = 9. Складываем все и получаем 111 весов!

```python
result = list(map(lambda x, y: x + y, [1, 2, 3], [4, 5, 6]))
print(result)
```

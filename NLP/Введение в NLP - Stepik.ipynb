Первая модель на основе **Transformers** это по сути **Google Translate** (начал показывать хорошие результаты перевода с 2016 года) и сейчас модель GPT-4 (также на основе архитектуры Transformers)

Но есть ограничения, например, оценить насколько эта задача хорошо выполнена или нет. Мы можем дать человеку задачу, он может ее разложить на подзадачи и начать выполнять. GPT так не может, она может просто генерировать текст, но хорошо или плохо она сгенерировала текст – непонятно.

**Embedding** – это векторное представление слова, которое кодирует его смысл при помощи чисел в каком-то пространстве, и в этом пространстве мы ожидаем, что слова с близким смыслом будут расположены рядом, а слова с разными смыслами далеко друг от друга.

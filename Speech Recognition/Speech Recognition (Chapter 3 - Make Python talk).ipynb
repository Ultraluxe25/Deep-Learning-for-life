```python
import speech_recognition as sr

class Robot:
    ''' Gives commands for current robot '''
    def __init__(self, name):
        self.name = name

    def do_order(self, order):
        ''' Actions for particular commands '''
        if order == 'stop':
            print(f'{self.name} is stopping')
        elif order == 'go ahead':
            print(f'{self.name} is going ahead')
        elif order == 'switch off':
            print(f'{self.name} is switching off...')
            print('Goodbye.', end=' ')
            return True
        else:
            print(f'{self.name} doesn\'t understand current command')

    def order_robot(self):
        ''' Robot is waiting for commands and then sleep '''
        speech = sr.Recognizer()
        print(f'{self.name} is waiting for commands...')

        with sr.Microphone() as source:
            # Switch on the microfone
            stop = False
            while not stop:
                speech.adjust_for_ambient_noise(source)
                audio = speech.listen(source)  # record audio with commands
                command = speech.recognize_google(audio)  # convert the audio into text
                stop = self.do_order(command)
                if not stop:
                    print(f'{self.name} is waiting for a new command')
        
        print(f'{self.name} is sleeping now')

        
if __name__ == '__main__':
    robot_name = 'Newman'
    newman = Robot(robot_name)
    newman.order_robot()
    print('---End---')
```
